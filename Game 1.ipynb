```python
def Guess_Process(in_number,rnd_number,life):
    Max_update=0
    Min_update=0
    if in_number > rnd_number:
        #print(">")
        guess_right = 0
        life = life -1
        Max_update = in_number-1
    elif in_number < rnd_number:
        #print("<")
        guess_right = 0
        life = life -1
        Min_update = in_number+1
    elif in_number == rnd_number:
        #print("=")
        guess_right = 1
    else:
        print("it is impossible")
        
    return guess_right, life, Max_update, Min_update
```
